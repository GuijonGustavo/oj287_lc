# DATE TIME

```python
import datetime
from miditime.miditime import MIDITime
```

```python
mymidi = MIDITime(80,'data_sonified.mid',7,3,4)
mymidi = MIDITime(80,obj+"_"+freq+_sonified.mid,7,3,4)
```

```python
def clean(x):
    year = int(x.split("-")[0])
    month = int(x.split("-")[1])
    day = int(x.split("-")[2])
    x = datetime.datetime(year,month,day)
    return x
```

```python
data['date'] = data['date'].apply(clean)
```

```python
my_data = data.to_dict('records')
#my_data
```

```python
my_data_epoched = [
    {
        'days_since_epoch': mymidi.days_since_epoch(d['date']),
        'Flux':d['Flux']
    } for d in my_data
]
```

```python
my_data_timed = [
    {
        'beat': mymidi.beat(d['days_since_epoch']),
        'Flux': d['Flux']
    } for d in my_data_epoched
]
```

```python
start_time = my_data_timed[0]['beat']
```

```python
def data_to_pitch_tuned(Flux):
    scale_pct = mymidi.linear_scale_pct(0,30,Flux)
    d_minor = ['D','E','F','G','A','Bd','C']
    note = mymidi.scale_to_note(scale_pct, d_minor)
    midi_pitch = mymidi.note_to_midi_pitch(note)
    #print(note)
    return midi_pitch
```

```python
note_list = []
for d in my_data_timed:
    note_list.append([
        d['beat'] - start_time,
        data_to_pitch_tuned(d['Flux']),
        90,
        2
    ])
```

```python
mymidi.add_track(note_list)
```

```python
mymidi.save_midi()
```
